```python
#Confidence interval adalah jenis statistik yang dihitung berdasarkan pengamatan
```

```python
import pandas as pd
import numpy as np
df= pd.read_csv('Heart.csv')
df
```

```python
df['Sex1'] = df.sex.replace({1:"Male", 0:"Female"})
```

```python
dx = df[["target", "Sex1"]].dropna()
```

```python
pd.crosstab(dx.target, dx.Sex1)
p_fm = 226/(86+226)
p_fm
```

```python
# 226 adalah peremuan penderita jantung, 86 adalah lakilaki, n = 226 + 86
```

```python
n = 86+226
n
```

```python
se_female = np.sqrt(p_fm * (1 - p_fm) / n)
se_female
```

```python
#Z_score adalah standard statistic (1.96) untuk menghitung confidence interval sebesar 95%/5% error

z_score = 1.96
lcb = p_fm - z_score * se_female #lower limit of the CI
ucb = p_fm + z_score * se_female #upper limit of the CI

lcb, ucb
```

```python
import statsmodels.api as sm
sm.stats.proportion_confint(n * p_fm, n)
```
